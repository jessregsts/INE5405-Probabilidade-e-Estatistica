```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv("br_inep_ideb_brasil.csv")
base = base[base['rede'] != 'total']
temp = base.select_dtypes(include='number')  # Mantém apenas colunas numéricas
```

```python
base.dtypes
```

```python
variaveis_para_analisar = ['taxa_aprovacao', 'indicador_rendimento', 'nota_saeb_matematica', 
                           'nota_saeb_lingua_portuguesa', 'nota_saeb_media_padronizada', 'ideb']

# Agrupar por ano e calcular a média das variáveis para cada ano
variacao_por_ano = base.groupby('ano')[variaveis_para_analisar].mean()

# Exibir a variação por ano
print("\nVariação por ano:")
print(variacao_por_ano)

# Plotar a variação de algumas variáveis ao longo dos anos
variacao_por_ano.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Variação das variáveis ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Valor')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
variacao_matematica_por_ano = base.groupby('ano')['nota_saeb_matematica'].mean()

# Exibir a variação da nota de matemática por ano
print("\nVariação da nota de matemática por ano:")
print(variacao_matematica_por_ano)

# Plotar a variação da nota de matemática ao longo dos anos
variacao_matematica_por_ano.plot(kind='line', figsize=(10, 6), marker='o', color='b')
plt.title('Variação da Nota SAEB de Matemática ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Nota SAEB de Matemática')
plt.grid(True)
plt.show()
```

```python
variacao_lingua_portuguesa_por_ano = base.groupby('ano')['nota_saeb_lingua_portuguesa'].mean()

# Exibir a variação da nota de portugues por ano
print("\nVariação da nota de matemática por ano:")
print(variacao_lingua_portuguesa_por_ano)

# Plotar a variação da nota de portugues ao longo dos anos
variacao_lingua_portuguesa_por_ano.plot(kind='line', figsize=(10, 6), marker='o', color='b')
plt.title('Variação da Nota SAEB de Língua Portuguesa ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Nota SAEB de Língua Portuguesa')
plt.grid(True)
plt.show()
```

```python
variacao_lingua_portuguesa_por_ano = base.groupby('ano')['ideb'].mean()

# Exibir a variação da nota de matemática por ano
print("\nVariação da media ao longo dos anos:")
print(variacao_lingua_portuguesa_por_ano)
plt.axhline(y=6, color='r', linestyle='--', label='Ideal (6)')
# Plotar a variação da nota de matemática ao longo dos anos
variacao_lingua_portuguesa_por_ano.plot(kind='line', figsize=(10, 6), marker='o', color='b')
plt.title('Variação da Média ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('Média Ideb')
plt.grid(True)
plt.show()
```

```python
# Agrupar por ano e calcular a média da taxa de aprovação e a projeção
taxa_de_aprovacao_ano = base.groupby('ano')['ideb'].mean()
projecao_ano = base.groupby('ano')['projecao'].mean()

# Exibir a variação da taxa de aprovação e da projeção por ano
print("\nVariação da taxa do ideb por ano:")
print(taxa_de_aprovacao_ano)

print("\nVariação da projeção por ano:")
print(projecao_ano)

# Plotar a variação da taxa de aprovação e da projeção ao longo dos anos
plt.figure(figsize=(10, 6))
plt.plot(taxa_de_aprovacao_ano, marker='o', color='b', label='Média Ideb')
plt.plot(projecao_ano, marker='o', color='g', label='Projeção')

# Título e rótulos
plt.title('Variação da Média Ideb e Projeção por Ano')
plt.xlabel('Ano')
plt.ylabel('Média (%)')
plt.grid(True)

# Adicionar a legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
